```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("titanic.txt")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df = df.drop_duplicates()
df.shape
```

```python
df['Pclass'].value_counts().plot(kind='bar')
plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
df['Age'].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Observations
- Most passengers belonged to 3rd class.
- Majority of passengers were between 20–40 years old.
- Missing values and duplicate rows were removed during data cleaning.
- Data visualization helped identify passenger distribution patterns.
